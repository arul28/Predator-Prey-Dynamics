```
import matplotlib.pyplot as plt
import numpy as np
```

```
def rk4(d, f, h): #this is a modified rk4 function without state vectors
 f1 = h*lv(d, f)
 f2 = h*lv(d+(f1/2),f+(h/2))
 f3 = h*lv(d+(f2/2),f+(h/2))
 f4 = h*lv(d+f3,f+h)
 return (f1+(2*f2)+(2*f3)+f4)/6
```

I wanted to see if I could graph the lotka volterra equations where both populations reach equilibrium. Equilibrium occurs when the derivative equals 0. So I set dx/dt and dy/dt equal to zero. I found that equilibrium (without extinction) will occur when x = c/(e*b) & y = (a/b).The way I coded my RK4 functions, I couldn't figure out how to input those functions. So I receated a simple Lotka Volterra function without RK4, since this was just to test what equilibrium would look like.

```
h = 0.00001 #I used a really tiny timestep, just to make sure I don't overshoot
xvals = [] #empty arrays for x values
yvals = [] #empty arrays for y values
#defining a,b,c,e,x,y
a = 1
b = 0.5
c = 0.5
e = 0.5
x = c/(e*b)
y = (a/b)

for i in range(1000): #I knew that if my function was going to work, 1000 would end up being my xmax, which is why I chose that value and didn't define an xmax
 xvals.append(x) #adding values to x array
 yvals.append(y) #addign values to y array
 dxdt = (a*x)-(b*x*y) #defining dx/dt
 dydt = (e*b*x*y)-(c*y) #defining dy/dt
 x += h*dxdt #setting x & y equal to the next correct value
 y += h*dydt

#plotting with the correct labels and legend
plt.xlabel("Time")
plt.ylabel("Population")
plt.plot(xvals, label="Prey Population")
plt.plot(yvals, label="Predator Population")
plt.legend(loc='upper right')
plt.axis([0,1000,-500,500])
plt.show()
```

```
h = 0.00001 #I used a really tiny timestep, just to make sure I don't overshoot
xvals = [] #empty arrays for x values
yvals = [] #empty arrays for y values
#defining a,b,c,e,x,y
a = 1
b = 0.5
c = 0.5
e = 0.5
x = c/e
y = (a/b)

for i in range(1000): #I knew that if my function was going to work, 1000 would end up being my xmax, which is why I chose that value and didn't define an xmax
 xvals.append(x) #adding values to x array
 yvals.append(y) #addign values to y array
 dxdt = (a*x)-(b*x*y) #defining dx/dt
 dydt = (e*x*y)-(c*y) #defining dy/dt
 x += h*dxdt #setting x & y equal to the next correct value
 y += h*dydt

#plotting with the correct labels and legend
plt.xlabel("Time")
plt.ylabel("Population")
plt.plot(xvals, label="Prey Population")
plt.plot(yvals, label="Predator Population")
plt.legend(loc='upper right')
plt.show()
```
